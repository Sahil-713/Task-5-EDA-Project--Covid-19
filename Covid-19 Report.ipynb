```python
df['sex_label'] = df['SEX'].map({1: 'Female', 2: 'Male'})
# Pie chart for gender distribution
df['sex_label'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'blue'])
plt.title('Gender Distribution')
plt.ylabel('')  # Hide y-axis label
plt.show()
```

```python
### Observation: Gender Distribution of COVID-19 Patients
'''
- The pie chart shows the gender distribution of COVID-19 patients based on the `SEX` field (1 = Female, 2 = Male).
- Females form a slightly higher proportion (about 50.1%) compared to males (49.9%) in the dataset.
- The gender distribution is almost balanced between males and females.
- Further analysis could be conducted to study if gender has any impact on hospitalization or mortality rates.'''
```

```python
sns.histplot(df['AGE'], bins=30)
plt.title('Age Distribution of COVID-19 Patients')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
### Observation: Age Distribution of COVID-19 Patients
'''
- The majority of COVID-19 patients are concentrated between the ages of 20 and 60 years.
- There is a noticeable peak around the 30–50 years age range.
- Very few patients are extremely young (below 10 years) or very old (above 80 years).
- The dataset shows that middle-aged adults are the most common age group affected.'''
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Select only numeric columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Plot
plt.figure(figsize=(15,12))  # Increase figure size
sns.heatmap(numeric_df.corr(), 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm', 
            annot_kws={"size":10})
plt.title("Relationship Between COVID-19 Variables", fontsize=18)
plt.show()


```

```python
# Observation Correlation Analysis
'''
We selected only the numeric variables from the COVID-19 dataset and calculated their pairwise correlation coefficients. The heatmap visualizes the strength and direction of these relationships:
- **Dark Red Cells**: Strong positive correlation (both variables increase together)
- **Dark Blue Cells**: Strong negative correlation (one variable increases, the other decreases)
- **Light Cells**: Weak or no correlation.

**Observation:**
Total confirmed cases show a strong positive correlation with total deaths, indicating that as the number of cases rises, deaths also increase. Recovered cases and active cases show a moderate negative correlation, suggesting that as recoveries go up, active cases come down.'''

```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df=pd.read_csv(r"D:\COVID-19\Covid Data.csv")

# Create death_status
df['death_status'] = df['DATE_DIED'].apply(lambda x: 0 if x == '9999-99-99' else 1)

# Take small sample
sample_df = df.sample(n=100, random_state=42)

# Select very few features
small_features = ['AGE', 'PNEUMONIA', 'DIABETES', 'death_status']

# Plot
sns.pairplot(sample_df[small_features], hue='death_status', palette='coolwarm')
plt.suptitle('Pairplot of Age, Pneumonia, Diabetes vs Death Outcome (Sample of 100 Patients)', y=1.02)
plt.show()

```

```python
### Observation: Pairplot of Age, Pneumonia, Diabetes vs Death Outcome
'''
- Patients with pneumonia and diabetes show higher mortality compared to those without.
- Older age is associated with increased death probability.
- Most survivors are younger and without critical comorbidities.
- The pairplot reveals clear clustering of death cases around higher age and presence of pneumonia and diabetes.'''

```
